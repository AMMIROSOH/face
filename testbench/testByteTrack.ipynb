```python
from src.bytetrack.byte_tracker import BYTETracker
import numpy as np
```

```python
tracker = BYTETracker(args={"track_thresh": 0.5, "match_thresh": 0.8, "track_buffer": 30, "mot20": False })
```

```python
video_frames = [
    np.array([[100, 150, 200, 300, 0.9, 0], [400, 100, 500, 250, 0.85, 0]]),
    np.array([[100, 150, 200, 300, 0.9, 0], [400, 100, 500, 250, 0.85, 0]]),
    np.array([[105, 155, 205, 305, 0.88, 0], [405, 105, 505, 255, 0.83, 0]]),
    # ...
]
```

```python
for frame_id, detections in enumerate(video_frames):
    online_targets = tracker.update(detections, [640, 480], [640, 480])
    
    print(f"Frame {frame_id}:")
    for t in online_targets:
        tlwh = t.tlwh  # [x, y, w, h]
        track_id = t.track_id
        print(f"  ID {track_id}: box {tlwh}")
```

```python
import time
start = time.time()
for i in range(1000):
    tracker.update(video_frames[0], [640, 480], [640, 480])
end = time.time()
avg_time = (end - start) / 1000
print(avg_time*1000, "ms")
```
